### 1. Classification Problem (Predict Deal Success)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# Data preprocessing
features = ['pitcher_ask_amount', 'ask_equity', 'ask_valuation', 'total_sharks_invested']
X = df[features]
y = df['deal']
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Logistic Regression Model
clf = LogisticRegression()
clf.fit(X_train_scaled, y_train)
```

```python
# Predictions and evaluation
y_pred = clf.predict(X_test_scaled)
print(classification_report(y_test, y_pred))
```

### 2. Regression Problem (Predict Deal Amount)

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Filter only pitches that got a deal
successful_pitches = df[df['deal'] == 1]
```

```python
# Define features and target
X = successful_pitches[['pitcher_ask_amount', 'ask_valuation', 'ask_equity', 'total_sharks_invested']]
y = successful_pitches['deal_amount']
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Linear Regression Model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predictions and evaluation
y_pred = regressor.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))
```

### 3. Clustering (Grouping Pitches Based on Characteristics)

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Features for clustering
X = df[['pitcher_ask_amount', 'ask_valuation', 'ask_equity', 'total_sharks_invested']].fillna(0)
```

```python
# Elbow method to find the optimal number of clusters
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
# Plot Elbow curve
plt.plot(range(1, 10), inertia)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
# Apply KMeans with the optimal number of clusters (e.g., 3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)
df['cluster'].value_counts()
```

### 4. Multi Label Classification (Predict Which Sharks Will Invest)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Define the features and target labels (shark deals)
features = ['pitcher_ask_amount', 'ask_equity', 'ask_valuation', 'total_sharks_invested']
target_labels = ['ashneer_deal', 'anupam_deal', 'vineeta_deal', 'peyush_deal', 'namita_deal']
```

```python
X = df[features].fillna(0)
y = df[target_labels].fillna(0)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Random Forest with MultiOutputClassifier
forest = RandomForestClassifier(n_estimators=100, random_state=42)
multi_target_forest = MultiOutputClassifier(forest, n_jobs=-1)
multi_target_forest.fit(X_train, y_train)
```

```python
# Predictions and evaluation
y_pred = multi_target_forest.predict(X_test)
```

```python
# Evaluation metrics
print("Model trained successfully!\n")
for idx, label in enumerate(target_labels):
    print(f"Results for {label}:")
    print(classification_report(y_test[label], y_pred[:, idx]))
    print("Accuracy:", accuracy_score(y_test[label], y_pred[:, idx]))
    print("\n" + "-"*50 + "\n")
```

```python
# Optional: Print overall model accuracy
overall_accuracy = accuracy_score(y_test, y_pred)
print("Overall model accuracy:", overall_accuracy)
```

### 5. Valuation Prediction (Predict Ask Valuation)

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
```

```python
# Define features and target
X = df[['pitcher_ask_amount', 'ask_equity']].fillna(0)
y = df['ask_valuation'].fillna(0)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Ridge Regression Model
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
```

```python
# Predictions and evaluation
y_pred = ridge.predict(X_test)
print("MAE:", mean_absolute_error(y_test, y_pred))
```
